```python
import random


# - - - Monty Hall problem - simulation proves bigger chance to win, when contestant switch his choice. - - - - -


n = 0

Count_SW = 0
Count_not_SW = 0

WIN_not_SW = [0]
WIN_SW = [0]



# - - - - - Used if contestant choose to switch doors - - - - -

def switch_game():
    global switched_choice
    global Count_SW

    Count_SW += 1
    print("sw...")

    for x in doors:
        if x != first_choice:
            switched_choice = x
            print("Contestant switched choice to doors n.", switched_choice,)

    if switched_choice == winning_door:
        WIN_SW.append(1)
        print("Contestant switched and WON")

    else:
        WIN_SW.append(0)
        print("Contestant switched and LOSE")



# - - - - - Main - - - - -

def game():
    global doors
    global first_choice
    global winning_door
    global to_open

    to_open = []
    doors = [1, 2, 3]
    winning_door = doors[random.randint(0, 2)]
    print("\nWinning door are n.  ", winning_door)

    first_choice = doors[random.randint(0, 2)]
    print("Contestant choosed n.", first_choice)


    # - - - - - Quizman determine doors to open, opening - - - - -

    doors_to_open = []
    for x in doors:

        if (x != winning_door and x != first_choice):

            doors_to_open.append(x)

    print("Possible to open     ", doors_to_open)

    while len(doors)>2:
            to_open = random.choice(doors_to_open)
            print("Quizman opens door n.", to_open)
            doors.remove(to_open)
            print("Doors remaining after open", doors)
    else:
            print("opening finished...")


    # - - - - - Choose if switch - - - - -

    switch = random.choice([True, False])

    if switch:
        switch_game()

    else:
        global Count_not_SW
        Count_not_SW += 1
        print("Contestant did NOT switched his choice.")
        if first_choice == winning_door:
            WIN_not_SW.append(1)
            print("Contestant did't switched and WON")
        else:
            WIN_not_SW.append(0)
            print("Contestant did't switched and LOSE")


# - - - - - Starting program - - - - -

number_of_games = 100

while n < number_of_games:
    n += 1
    print("\n\nGame n: ", n)
    game()
    print(Count_SW, "games contestant swiched")
    print(Count_not_SW, "games contestant not-swiched\n-------------------" )


# - - - - - Printing results - - - - -

else:

    sw_perc = format((WIN_SW.count(1))/(Count_SW/100),'.2f')
    nsw_perc = format((WIN_not_SW.count(1))/(Count_not_SW/100),'.2f')


    print("\n\n====================================================================", "\nFinished...")
    print("\nGame has been simulated %s times." % number_of_games)

    print("\nContestant WON ", WIN_SW.count(1) ," of %s games when switched choice." % Count_SW, "... ........Probability to win", sw_perc, "%." )
    print("Contestant WON ", WIN_not_SW.count(1)," of %s games when did NOT switched choice." % Count_not_SW, "... Probability to win", nsw_perc, "%." )

    print("\nYou better switch!")
```
